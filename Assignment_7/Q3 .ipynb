```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./framingham.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['cigsPerDay'].fillna(df['cigsPerDay'].mean(), inplace =  True)
```

```python
df['BPMeds'].fillna(df['BPMeds'].mode()[0], inplace = True)
```

```python
df['totChol'].fillna(df['totChol'].mean(), inplace = True)
```

```python
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
```

```python
df['heartRate'].fillna(df['heartRate'].mean(),inplace = True)
```

```python
df['glucose'].fillna(df['glucose'].mean(),inplace=True)
```

```python
df.info()
```

```python
df.corr()
```

```python
y = df['TenYearCHD']
x = df.drop(['TenYearCHD'], axis = 1)
```

```python
# split into train and test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# create model

from sklearn.linear_model import LogisticRegressionCV

model = LogisticRegressionCV()

model.fit(x_train,y_train)
```

```python
# load model
y_true = y_test

y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# confusion matrix

from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score

print(f"Accuracy score : {accuracy_score(y_true,y_pred)*100 :.2f}%")

print(f"Precision Score : {precision_score(y_true,y_pred)*100 :.2f}%")

print(f"Recall score : {recall_score(y_true,y_pred)*100 : .2f}%")

print(f"F1 socre : {f1_score(y_true,y_pred)*100 :.2f}%")
```
